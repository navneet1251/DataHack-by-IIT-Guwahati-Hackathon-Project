```python
# we will perform the ordinal encoding for age_group, education, income_poverty features

age_group_map ={'18 - 34 Years':1,'35 - 44 Years':2, '45 - 54 Years':3, '55 - 64 Years':4, '65+ Years':5 }
education_map ={'< 12 Years':1,'12 Years':2, 'Some College':3, 'College Graduate':4 }
income_poverty_map ={'Below Poverty':1,'> $75,000':2, '<= $75,000, Above Poverty':3 }
```

```python
# now we will perform frequency or count encoding for the remaining categorical features

categorical_remaining_features = [ 'race', 'sex', 'marital_status', 'rent_or_own', 'employment_status', 'hhs_geo_region',
       'census_msa', 'employment_industry', 'employment_occupation']
```

```python
def map_value_counts(df, categorical_features):
    
    for feature in categorical_features:
        count_map = df[feature].value_counts().to_dict()
        df[f'{feature}_map'] = df[feature].map(count_map)
        
    return df
```

```python
# Spliting the data into training and testing sets

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
```

```python
df=pd.read_csv(r"C:\Users\NEHA1\Documents\Summer Analystics Assignments\Hackathon_1\dataset and all\training_set_labels.csv")
df
```

```python

data_imputed =pd.merge(data_imputed, df, on='respondent_id', how='left')
# this is showing error because we have already run this cell many times and duplicate formed, so we have to drop the duplicated columns  
```

```python
data_xyz=data_imputed['xyz_vaccine']
data_seasonal = data_imputed['seasonal_vaccine']
```

```python
# we will drop the non usefull columns
data_imputed.drop(['xyz_vaccine_x','seasonal_vaccine_x','xyz_vaccine_y','seasonal_vaccine_y','xyz_vaccine','seasonal_vaccine', 'respondent_id' ], axis=1, inplace=True)
```

```python
#Splitting the data into training and testing sets
data_train, data_test, data_xyz_train, data_xyz_test, data_seasonal_train, data_seasonal_test = train_test_split(data_imputed_scaled, data_xyz, data_seasonal, test_size=0.2, random_state=42)

# Initialize the models
model_xyz = RandomForestClassifier(random_state=42)
model_seasonal = RandomForestClassifier(random_state=42)
```

```python
# Training the models
model_xyz.fit(data_train, data_xyz_train)
model_seasonal.fit(data_train, data_seasonal_train)
```

```python
# Predict probabilities
data_xyz_pred_proba = model_xyz.predict_proba(data_test)[:, 1]
data_seasonal_pred_proba = model_seasonal.predict_proba(data_test)[:, 1]
```

```python
# Evaluation of the model using ROC AUC score

roc_auc_xyz = roc_auc_score(data_xyz_test, data_xyz_pred_proba)
roc_auc_seasonal = roc_auc_score(data_seasonal_test, data_seasonal_pred_proba)
mean_roc_auc = (roc_auc_xyz + roc_auc_seasonal) / 2

print(f"ROC AUC for xyz_vaccine: {roc_auc_xyz}")
print(f"ROC AUC for seasonal_vaccine: {roc_auc_seasonal}")
print(f"Mean ROC AUC: {mean_roc_auc}")
```

```python
# predicted probabilities on the full dataset

data_full_scaled = scaler.transform(data_imputed)

data_xyz_pred_full_proba = model_xyz.predict_proba(data_full_scaled)[:, 1]
data_seasonal_pred_full_proba = model_seasonal.predict_proba(data_full_scaled)[:, 1]


# preparing  the submission file in the given format
submission = pd.DataFrame({
    'respondent_id': data['respondent_id'],
    'xyz_vaccine': data_xyz_pred_full_proba,
    'seasonal_vaccine': data_seasonal_pred_full_proba
})

submission.to_csv('submission.csv', index=False)
```
